```python
# 3_B
n=0
while(True):
    var=int(input())
    n+=1
    if(var!=0):
        print("Case "+str(n)+": "+str(var))
    elif(int(var)==0):
        break
```

```python
# 3_C
while(True):
    var=input()
    var=var.split()
    x=int(var[0])
    y=int(var[1])
    if(x==0 and y==0):
        break
    else:
        if(x>y):
            print(str(y)+" "+str(x))
        else:
            print(str(x)+" "+str(y))
```

```python
# 3_D
var=input()
var=var.split()
a=int(var[0])
b=int(var[1])
c=int(var[2])
count=0
for i in range(a,(b+1)):
    if(c%i==0):
        count+=1
    else:
        count=count
print(count)
```
